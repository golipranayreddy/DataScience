# 1.find the 20 most popular movies with a rank more than 8.0

```python
import pandas as pd
import numpy as np
```

```python
ratings = pd.read_csv("D:/DataScience/DADV/Assignment-1/Data/Ratings.tsv", delimiter = '\t')
```

```python
ratings.head()
```

```python
basics_movie = pd.read_csv("D:/DataScience/DADV/Assignment-1/Data/Title_Basics.tsv", delimiter = '\t')
```

```python
basics_movie.head()
```

```python
top_ratings = ratings[ratings['averageRating'] > 8.0]
top_ratings
```

```python
sorted_top_ratings = top_ratings.sort_values('numVotes', ascending = False).head(20)
sorted_top_ratings
```

```python
merged_rating_and_basics = pd.merge(sorted_top_ratings, basics_movie, on= 'tconst')
merged_rating_and_basics
```

```python
merged_rating_and_basics_filter_by_movie = merged_rating_and_basics.loc[merged_rating_and_basics['titleType'] == 'movie']
```

```python
top_20_movie_details = merged_rating_and_basics_filter_by_movie[['originalTitle','averageRating' ]].head(20)
top_20_movie_details
```

```python
top_20_movie_details =top_20_movie_details.sort_values('averageRating', ascending = False)
top_20_movie_details
```

```python
top_20_movie_details.to_csv('D:/DataScience/DADV/Assignment-1/Popular_Movies.csv' )
```

# 2.	find the 20 best rated movies with over 40,000 votes in the 2000s (year >= 2000)

```python
basics_movie.dtypes
```

```python
basics_movie['startYear'] = pd.to_numeric(basics_movie['startYear'], errors='coerce')
```

```python
basics_movie.dtypes
```

```python
basics_year_2000 = basics_movie[basics_movie.startYear >= 2000]
basics_year_2000
```

```python
merged_basics_year_2000_rating = pd.merge(basics_year_2000, ratings, on= 'tconst')
merged_basics_year_2000_rating
```

```python
basics_year_2000_rating_min_40000_votes = merged_basics_year_2000_rating[merged_basics_year_2000_rating['numVotes'] > 40000]
basics_year_2000_rating_min_40000_votes
```

```python
sorted_basics_year_2000_rating_min_40000_votes = basics_year_2000_rating_min_40000_votes.sort_values('averageRating', ascending = False)
sorted_basics_year_2000_rating_min_40000_votes
```

```python
min_40000_votes_only_movies = sorted_basics_year_2000_rating_min_40000_votes.loc[sorted_basics_year_2000_rating_min_40000_votes['titleType'] == 'movie']
min_40000_votes_only_movies
```

```python
Top_rated_movies_min_40000= min_40000_votes_only_movies[['originalTitle','numVotes', 'startYear','averageRating' ]].head(20)
Top_rated_movies_min_40000
```

```python
Top_rated_movies_min_40000.to_csv('D:/DataScience/DADV/Assignment-1/Movie_Highvotes.csv')
```
