```python
def daily_gain_or_loss(symbol):
    df_daily = pd.read_csv("D:/DataScience/DADV/DADV EXAM/Daily Scrape Files/" + symbol + ".csv")
    df_daily["Gain_Or_Loss"] = ((df_daily["Close"] - df_daily["Close"].shift(1)) / df_daily["Close"]) * 100
    daily_df_dict[symbol] = df_daily
    daily_max_gain_or_loss_dict[symbol] = df_daily["Gain_Or_Loss"].sum()
    daily_gain_or_loss_dict[symbol] = df_daily
#     print(daily_gain_or_loss_dict)
```

```python
def weekly_gain_or_loss(symbol):    
    df_weekly = pd.read_csv("D:/DataScience/DADV/DADV EXAM/Weekly Scrape Files/" + symbol + ".csv")
    df_weekly["Gain_Or_Loss"] = ((df_weekly["Close"] - df_weekly["Close"].shift(1)) / df_weekly["Close"]) * 100
    weekly_df_dict[symbol] = df_weekly
    weekly_max_gain_or_loss_dict[symbol] = df_weekly["Gain_Or_Loss"].sum()
    weekly_gain_or_loss_dict[symbol] = df_weekly
#     print(weekly_gain_or_loss_dict)
```

```python
def monthly_gain_or_loss(symbol):
    df_monthly = pd.read_csv("D:/DataScience/DADV/DADV EXAM/Monthly Scrape Files/" + symbol + ".csv")
    df_monthly["Gain_Or_Loss"] = ((df_monthly["Close"] - df_monthly["Close"].shift(1)) / df_monthly["Close"]) * 100
    monthly_df_dict[symbol] = df_monthly
    monthly_max_gain_or_loss_dict[symbol] = df_monthly["Gain_Or_Loss"].sum()
    monthly_gain_or_loss_dict[symbol] = df_monthly
#     print(monthly_gain_or_loss_dict)
```
